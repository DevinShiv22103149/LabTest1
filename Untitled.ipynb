```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
City1 = np.array([100,120.85,90,110,95])
City2 = np.array([80,75,60,95,85,90])
City3 = np.array([150,140,135,160,155,170])
```

```python
def totalrainfall(data:int):
    sum1 = 0
    for i in data:
        sum1 = sum1 + i
    print(sum1)
```

```python
totalrainfall(City1)
```

```python
totalrainfall(City2)
totalrainfall(City3)
```

```python
def average(data:int):
    sum1 = 0
    count = 0
    for i in data:
        sum1 = sum1 + i
        count += 1
    sum1 = sum1/count
    return sum1
```

```python
av1 = average(City1)
print(av1)
```

```python
av2 = average(City2)
print(av2)
av3 = average(City3)
print(av3)
```

```python
def totalaverage1(data1:int,data2:int,data3:int):
    totalav = (data1+data2+data3)/3
    print(totalav)
```

```python
totalaverage1(av1,av2,av3)
```

```python
def plotall(data1:int,string:str,data2:int,data3:int):
    plt.plot(data1,color='Red',zorder=5)
    plt.plot(data2,color='Blue',zorder=5)
    plt.plot(data3,color='Green',zorder=5)
    plt.title("Cities")
    plt.xlabel("Month")
    plt.ylabel("Rainfall")
    
    plt.show()
```

```python
plotall(City1,"City1",City2,City3)
```

```python
def maxmin(data:int):
    max1 = np.max(data)
    min1 = np.min(data)
    da = max1 - min1
    print(da)
```

```python
da1 = maxmin(City1)
```

```python
da2 = maxmin(City2)
da3 = maxmin(City3)
```

```python
plt.bar(da1,da2,da3,height=5)
```
